```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
from matplotlib import pyplot as plt
%matplotlib inline

# helpful character encoding module
import chardet
```

```python
data = pd.read_csv("DSI_kickstarterscrape_dataset_cleaned (4).csv", engine='python')
data
```

```python
data.describe()
```

```python
plt.figure(figsize=(16,10))
sns.countplot(x='category', data=data)
```

```python
sns.countplot(x='year', data=data)
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(x="category", y="pledged", data=data, palette='rainbow') 
plt.xlim(-1, 15)
plt.ylim(-1000, 10000)
```

```python
plt.figure(figsize=(16,10))
sns.stripplot(x="category", y="pledged", data=data)
```

```python
plt.figure(figsize=(16,10))
sns.pointplot(x="category", y="funded_percentage", data=data)
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x="goal", y="backers", data=data)
plt.xlim(-10, 100)
plt.ylim(-100, 1000)
```

```python
plt.figure(figsize=(11,7))
sns.lmplot(x='goal',y='backers',data=data,col='status',palette='Set1')
```
